# PyCity Schools Analysis

In the data below schools were separated into 2 types. Charter schools did better than district schools in measures of average reading and math scores, as well as the pecent of their students who passed reading and math. Schools with larger student bodies showed less students passing reading and math and lower average scores for reading and math. Hernandez High School had one of the largest budgets but had the second lowest overall passing percentage.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# reset the pandas index in a new dataframe
school_types = school_data.set_index(["school_name"])
school_types
```

```python
# Calculate the total student count per school
studPerSchool = school_data_complete["school_name"].value_counts()
studPerSchool
```

```python
# avg of budget per school, groupby on school_data_complete
groupedSchoolDF = school_data_complete.groupby("school_name")
budgetPerSchool = groupedSchoolDF.mean("budget")
# turn budget into series
budgetPerSchool = budgetPerSchool["budget"]
budgetPerSchool
```

```python
# Calculate the total school budget and per capita spending per school (school bud / stud count)
capitaPerSchool = budgetPerSchool / studPerSchool
capitaPerSchool
```

```python
# avg test scores
groupedSchoolDF = school_data_complete.groupby("school_name")
# avg math scores
avgMathScoreDF = groupedSchoolDF.mean("math_score")
# covert to series
mathPerSchool = avgMathScoreDF["math_score"]
mathPerSchool
```

```python
# avg test scores
groupedSchoolDF = school_data_complete.groupby("school_name")
# avg reading scores
avgReadingScoreDF = groupedSchoolDF.mean("reading_score")
# covert to series
readingPerSchool = avgReadingScoreDF["reading_score"]
readingPerSchool
```

```python
#  Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
school_students_passing_math = students_passing_math.groupby("school_name")
passMathPerSchool = school_students_passing_math["student_name"].count()
passMathPerSchool
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
school_students_passing_reading = students_passing_reading.groupby("school_name")
passReadingPerSchool = school_students_passing_reading["student_name"].count()
passReadingPerSchool
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_pass = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
school_students_pass = students_pass.groupby(["school_name"]).size()
school_students_pass
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = passMathPerSchool / studPerSchool * 100
per_school_passing_reading = passReadingPerSchool / studPerSchool * 100
overall_passing_rate = school_students_pass / studPerSchool * 100
overall_passing_rate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_sum = pd.DataFrame(
    {
        "School Type": school_type,
        "Total Students": studPerSchool,
        "Total School Budget": budgetPerSchool,
        "Budget per Student": capitaPerSchool,
        "Avg Math Score": mathPerSchool,
        "Avg Reading Score": readingPerSchool,
        "% Pass Math": per_school_passing_math,
        "% Pass Reading": per_school_passing_reading,
        "% Overall Pass": overall_passing_rate,        
    }
)

# Formatting
per_school_sum["Total Students"] = per_school_sum["Total Students"].map("{:,}".format)
per_school_sum["Total School Budget"] = per_school_sum["Total School Budget"].map("${:,.2f}".format)
per_school_sum["Budget per Student"] = per_school_sum["Budget per Student"].map("${:,.2f}".format)
per_school_sum["Avg Math Score"] = per_school_sum["Avg Math Score"].map("{:.2f}".format)
per_school_sum["Avg Reading Score"] = per_school_sum["Avg Reading Score"].map("{:.2f}".format)
per_school_sum["% Pass Math"] = per_school_sum["% Pass Math"].map("{:.2f}%".format)
per_school_sum["% Pass Reading"] = per_school_sum["% Pass Reading"].map("{:.2f}%".format)
per_school_sum["% Overall Pass"] = per_school_sum["% Overall Pass"].map("{:.2f}%".format)

# Remove school name from index
per_school_sum.index.name = None

# Display the DataFrame
per_school_sum
```

## Scores by School Spending

```python
#display school sum df
per_school_sum.head()
```

```python
#display max and min to determine ranges
print(per_school_sum["Budget per Student"].max())
print(per_school_sum["Budget per Student"].min())
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_sum.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(capitaPerSchool, spending_bins, labels=labels)
school_spending_df.index.name = None
school_spending_df
```

```python
# return strings from mapping back to floats
school_spending_df["Avg Math Score"] = school_spending_df["Avg Math Score"].astype(float)
school_spending_df["Avg Reading Score"] = school_spending_df["Avg Reading Score"].astype(float)
school_spending_df["% Pass Math"] = school_spending_df["% Pass Math"].str.replace("%","").astype(float)
school_spending_df["% Pass Reading"] = school_spending_df["% Pass Reading"].str.replace("%","").astype(float)
school_spending_df["% Overall Pass"] = school_spending_df["% Overall Pass"].str.replace("%","").astype(float)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Pass Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Pass Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Pass"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame(
    {
        "Avg Math Score": spending_math_scores,
        "Avg Reading Score": spending_reading_scores,
        "% Pass Math": spending_passing_math,
        "% Pass Reading": spending_passing_reading,
        "% Overall Pass": overall_passing_spending,
    }
)

#Formatting
spending_summary["Avg Math Score"] = spending_summary["Avg Math Score"].map("{:.2f}".format)
spending_summary["Avg Reading Score"] = spending_summary["Avg Reading Score"].map("{:.2f}".format)
spending_summary["% Pass Math"] = spending_summary["% Pass Math"].map("{:.2f}%".format)
spending_summary["% Pass Reading"] = spending_summary["% Pass Reading"].map("{:.2f}%".format)
spending_summary["% Overall Pass"] = spending_summary["% Overall Pass"].map("{:.2f}%".format)

# Display results
spending_summary
```

## Scores by School Size

```python
#create copy
school_spending_df2 = per_school_sum.copy()
```

```python
# convert total student column
school_spending_df2["Total Students"] = school_spending_df["Total Students"].str.replace(",","").astype(int)
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (less than 1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

school_spending_df2["School Size"] = pd.cut(school_spending_df2["Total Students"], size_bins, labels=labels)
school_spending_df2.head()
```

```python
# return strings from mapping back to floats
school_spending_df2["Avg Math Score"] = school_spending_df2["Avg Math Score"].astype(float)
school_spending_df2["Avg Reading Score"] = school_spending_df2["Avg Reading Score"].astype(float)
school_spending_df2["% Pass Math"] = school_spending_df2["% Pass Math"].str.replace("%","").astype(float)
school_spending_df2["% Pass Reading"] = school_spending_df2["% Pass Reading"].str.replace("%","").astype(float)
school_spending_df2["% Overall Pass"] = school_spending_df2["% Overall Pass"].str.replace("%","").astype(float)
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_spending_df2.groupby(["School Size"])["Avg Math Score"].mean()
size_reading_scores = school_spending_df2.groupby(["School Size"])["Avg Reading Score"].mean()
size_passing_math = school_spending_df2.groupby(["School Size"])["% Pass Math"].mean()
size_passing_reading = school_spending_df2.groupby(["School Size"])["% Pass Reading"].mean()
size_overall_passing = school_spending_df2.groupby(["School Size"])["% Overall Pass"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame(
    {
        "Avg Math Score": size_math_scores,
        "Avg Reading Score": size_reading_scores,
        "% Pass Math": size_passing_math,
        "% Pass Reading": size_passing_reading,
        "% Overall Pass": size_overall_passing,
    }
)

#Formatting
size_summary["Avg Math Score"] = size_summary["Avg Math Score"].map("{:.2f}".format)
size_summary["Avg Reading Score"] = size_summary["Avg Reading Score"].map("{:.2f}".format)
size_summary["% Pass Math"] = size_summary["% Pass Math"].map("{:.2f}%".format)
size_summary["% Pass Reading"] = size_summary["% Pass Reading"].map("{:.2f}%".format)
size_summary["% Overall Pass"] = size_summary["% Overall Pass"].map("{:.2f}%".format)

# Display results
size_summary
```
